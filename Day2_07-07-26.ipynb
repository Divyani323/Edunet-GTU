```python
print("LIST")
l1 = ["Priya","Radhe","Sandy", "Happy"]
l2 = [30,14,27,36,2]
l1.append(10)              # 1. append()
l1.insert(1, "Amol")       # 2. insert()
l1.remove("Sandy")         # 3. remove()
l2.sort()                  # 4. sort()
l1.pop()                   # 5. pop()
print("List:", l1)

print("\nTUPLE")
a1 = (10, 20, 30, 20, 40)
print("Count of 20:", a1.count(20))     # 1. count()
print("Index of 30:", a1.index(30))     # 2. index()
print("Length:", len(a1))               # 3. len()
print("Maximum:", max(a1))              # 4. max()
print("Minimum:", min(a1))              # 5. min()

print("\nSET")
colors = {"Red", "Blue", "Green"}

colors.add("Yellow")              # 1. add()
colors.remove("Blue")             # 2. remove()
colors.update({"Black", "White"}) # 3. update()
colors.discard("Pink")            # 4. discard()
copy_set = colors.copy()          # 5. copy()

print("Set:", colors)

print("\n DICTIONARY ")
student = {
    "Name": "Divyani",
    "Age": 21,
    "Course": "Python" 
}

print(student.keys())      # 1. keys()
print(student.values())    # 2. values()
print(student.items())     # 3. items()
print(student.get("Age"))  # 4. get()
student.update({"City": "Ahmedabad"})  # 5. update()
print(student)

print("\nINPUT")
name = input("Enter your name: ")
print("Hello,", name)

print("\n IF ")
num = 10
if num > 0:
    print("Positive Number")
    
print("\n IF ELSE ")
age = 17
if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")

print("\n IF ELIF ELSE ")
marks = 75
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")

print("\n NESTED IF ELSE ")
number = 12
if number > 0:
    if number % 2 == 0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Negative Number")


print("\n FOR LOOP ")
for i in range(1, 6):
    print(i)

print("\n WHILE LOOP ")
count = 1
while count <= 5:
    print(count)
    count += 1

print("\n BREAK ")
for i in range(1, 10):
    if i == 5:
        break
    print(i)

print("\n CONTINUE ")
for i in range(1, 6):
    if i == 3:
        continue
    print(i)

print("\n PASS ")
for i in range(1, 4):
    if i == 2:
        pass
    print(i)

print("\n RANGE FUNCTION ")
print(list(range(1, 6)))

print("\n LEN FUNCTION ")
print("Length of list:", len(l1))


print("\n TYPE FUNCTION ")
print(type(l1))
print(type(a1))
print(type(colors))
print(type(student))
```
