```python
import pandas as pd
import numpy as np
df = pd.read_csv("green.csv")
print("First Five Records")
print(df.head())
print("\nDataset Information")
print(df.info())

print("\nShape of Dataset")
print(df.shape)

print("\nColumn Names")
print(df.columns)

print("\nStatistical Summary")
print(df.describe(include='all'))

print("\nMissing Values")
print(df.isnull().sum())
print("\nHandling Missing Values...")
df["Training_Hours"] = df["Training_Hours"].fillna(
 df["Training_Hours"].mean())

df["Sustainability_Score"] = df["Sustainability_Score"].fillna(
    df["Sustainability_Score"].mean())
df["Skill_Level"] = df["Skill_Level"].fillna(
    df["Skill_Level"].mode()[0])

print("\nMissing Values After Cleaning")
print(df.isnull().sum())
print("\nAverage Sustainability Score by Skill Level")
print(df.groupby("Skill_Level")["Sustainability_Score"].mean())

print("\nAverage Training Hours by Education")
print(df.groupby("Education")["Training_Hours"].mean())

print("\nAverage Sustainability Score by Employment Status")
print(df.groupby("Employment_Status")["Sustainability_Score"].mean())

print("\nAverage Sustainability Score by Gender")
print(df.groupby("Gender")["Sustainability_Score"].mean())
print("\nCorrelation Matrix")

print(df[["Age",
          "Training_Hours",
          "Sustainability_Score"]].corr())

print("\nCleaned Dataset")
print(df)

print("\nAnalysis Completed Successfully.")
```
